<hr>
<p style="color:#2997DA"> 3. What are the top 5 establishments with a <code>RatingValue</code> rating value of <font style="color:#780325"><b>'5'</b></font>, sorted by lowest hygiene score, nearest to the new restaurant added, <font style="color:#780325"><b>"Penang Flavours"</b></font>?</p>
<ul style="color:green">
    <li>Search within 0.01 degree on either side of the latitude and longitude.</li>
    <li>Rating value must equal 5</li>
    <li>Sort by hygiene score</li>
    <li>Print the results</li>
    <li>Convert result to Pandas DataFrame</li>
</ul>

```python
degree_search = 0.01
latitude      = establishments.find_one({"BusinessName" : "Penang Flavours"})["geocode"]["latitude"]
longitude     = establishments.find_one({"BusinessName" : "Penang Flavours"})["geocode"]["longitude"]

add=(set_param(degree_search,[latitude, longitude]))[0]
sub=(set_param(degree_search,[latitude, longitude]))[1]
```

```python
look_query=Queries(establishments,[{'RatingValue' : {'$regex' : '5'},\
                                       'geocode.latitude' : {'$lte': add[0], '$gte': sub[0]},\
                                       'geocode.longitude': {'$lte': add[1], '$gte': sub[1]}}])
look_query.sort_limit('scores.Hygiene', 1, 5)
```

```python
look_query.ConvertToDF()
```

<hr>
<p style="color:#2997DA"> 4. How many establishments in each Local Authority area have a hygiene score of 0?</p>
<ul style="color:green">
    <li>Create a pipeline that:</li>
    <ol>
        <li>Matches establishments with a hygiene score of 0</li>
        <li>Groups the matches by Local Authority</li>
        <li>Sorts the matches from highest to lowest</li>
    </ol>
    <li>Print the number of documents in the result and the first 10 results</li>
    <li>Convert the result to a Pandas DataFrame, Display the number of rows, and 10 first rows</li>
</ul>

```python
pipline_query=Queries(establishments,[{'$match':{'scores.Hygiene':{"$eq":0}}},\
                                      {'$group':{'_id':'$LocalAuthorityName','count':{'$sum': 1}}},\
                                      {'$sort' :{'count':-1}}])
pipline_query.set_result()
pipline_query.get_result(10)
```

```python
pipline_query.ConvertToDF()
```
